# Modelos predictivos 


## Regresión lineal

Simulación básica para estudiar la regresion lineal

```python
# librerias necesarias

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

# Simular los datos



```python
np.random.seed(123)
X = 2 * np.random.rand(1000, 1) # parametro X en 0 y 2
print(X.mean())
print(X.min())
print(X.max())

Y = 4 + 5 * X*X + np.random.randn(1000, 1)  # y = 4 + 3x + ruido

model= LinearRegression()
model.fit(X,Y)

# Mostrar parámetros del modelo
print(f"Pendiente (a): {model.coef_[0][0]:.2f}")  # Coeficiente de x
print(f"Intercepto (b): {model.intercept_[0]:.2f}")  # Valor de y cuando x=0

model.predict(X)
```

```python

# Graficar datos y línea de ajuste
plt.scatter(X, Y, label='Datos reales', alpha=0.7)
plt.plot(X, model.predict(X), color='orange', linewidth=2, label='Línea ajustada')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Lineal: Datos vs. Modelo')
plt.legend()
plt.show()
```

# REGRESIÓN LOGÍSTICA

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
# Crear dataset sintético para clasificación
X, y = make_classification(
    n_samples=1000, n_features=2, n_redundant=0,
    n_clusters_per_class=1, random_state=0
)
```

```python
# Entrenar modelo de regresión logística
clf = LogisticRegression()
clf.fit(X, y)
```

```python
# Generar malla para visualizar frontera de decisión
grid = np.linspace(X.min()-1, X.max()+1, 200)
xx, yy = np.meshgrid(grid, grid)
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

```

```python
# Dibujar resultado
plt.scatter(X[:, 0], X[:, 1], label='Datos')
plt.title('Frontera de Decisión: Regresión Logística')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# Dibujar resultado
plt.contourf(xx, yy, Z, alpha=0.2)  # Zonas clasificadas
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', label='Datos')
plt.title('Frontera de Decisión: Regresión Logística')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

# CLUSTERING CON K-MEANS

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Generar datos sintéticos con 3 concentraciones
data = np.vstack([
    np.random.randn(1000, 2) + shift
    for shift in [(0, 0), (5, 5), (0, 5), (8, 9)]
])

data
```

```python
# Entrenar KMeans con 3 clusters
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(data)

labels = kmeans.labels_        # Etiqueta de cluster para cada punto
centers = kmeans.cluster_centers_  # Coordenadas de los centroides
```

```python
# Visualizar clusters y centroides
plt.scatter(data[:,0], data[:,1], alpha=0.6, edgecolor='k')
plt.title('Clustering K-Means: 3 Grupos')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
# Visualizar clusters y centroides
plt.scatter(data[:,0], data[:,1], c=labels, alpha=0.6, edgecolor='k')
plt.scatter(centers[:,0], centers[:,1],
            marker='X', s=200, linewidths=2, edgecolor='r',
            label='Centroides')
plt.title('Clustering K-Means: 3 Grupos')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
